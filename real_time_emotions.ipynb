```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the pre-trained model
model = load_model('emotion_model_best .h5')  # Replace with your model's path

# Define the emotion labels (ensure these match the labels used in training your model)
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

# Load the pre-trained face detector from OpenCV (Haar Cascade or any other method)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize the webcam
cap = cv2.VideoCapture(0)

# Continuously capture frames from the webcam
while True:
    ret, frame = cap.read()
    
    if not ret:
        print("Failed to capture frame")
        break

    # Convert the frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # Extract the face region from the frame
        face = frame[y:y+h, x:x+w]
        
        # Preprocess the face image for the model
        # Convert the face to grayscale (if the model expects grayscale input)
        face_gray = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
        
        # Resize the face image to match the model's input size (e.g., 48x48 for many emotion models)
        face_resized = cv2.resize(face_gray, (48, 48))
        
        # Normalize the image (if required by your model)
        face_resized = face_resized / 255.0  # Normalize pixel values to [0, 1]
        
        # Reshape the image to the model's expected input shape (e.g., (1, 48, 48, 1) for a grayscale model)
        face_input = np.expand_dims(face_resized, axis=-1)  # Add channel dimension
        face_input = np.expand_dims(face_input, axis=0)  # Add batch dimension
        
        # Make the prediction using the model
        emotion_probs = model.predict(face_input)
        predicted_emotion_index = np.argmax(emotion_probs)
        predicted_emotion = emotion_labels[predicted_emotion_index]
        
        # Draw a rectangle around the face and display the emotion
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame, predicted_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    # Show the frame with the face and emotion label
    cv2.imshow('Emotion Detection', frame)

    # Exit the loop if the user presses the 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
